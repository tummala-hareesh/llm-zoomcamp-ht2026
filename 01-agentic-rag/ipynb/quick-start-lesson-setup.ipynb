## Check Open AI client

```python
from dotenv import load_dotenv
load_dotenv()

from openai import OpenAI
openai_client = OpenAI()
```

## Interact with LLM (lack our data)

```python
def llm(prompt):
    response = openai_client.responses.create(
        model="gpt-5.4-mini",
        input=prompt
    )
    return response.output_text
```

```python
llm("Hey, what's up?")
```

```python
question = "I just discovered the course. Can I join now? Can I join late?"
answer = llm(question)
print(answer)
```

```python
question = "In 1 paragraph, tell me how to cook salmon."
answer = llm(question)
print(answer)
```

```python
question = "What are the alternatives to RestoraLax in India for 4 yr olds? I'm in Almasugda, Hyderabad. Give a summary of closest pharmacies that I can go to and find the product you recommend."
answer = llm(question)
print(answer)
```
